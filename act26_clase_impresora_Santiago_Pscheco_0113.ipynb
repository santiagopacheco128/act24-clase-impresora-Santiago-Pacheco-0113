```python
# clase impresora marca, tipo, velocidad
# funciones, leerdatos(), mostrardatos()
# crar 2 objetos impresora1, impresoara2

class Impresora:
    def __init__(self, marca, tipo, velocidad):
        self.marca = marca
        self.tipo = tipo
        self.velocidad = velocidad

    def leerdatos(self):
        self.marca = input("Ingrese la marca de la impresora: ")
        self.tipo = input("Ingrese el tipo (Inyección/Láser): ")
        self.velocidad = input("Ingrese la velocidad (ppm): ")

    def mostrardatos(self):
        print("\n--- Datos de la Impresora ---")
        print("Marca:     ", self.marca)
        print("Tipo:      ", self.tipo)
        print("Velocidad: ", self.velocidad, "ppm")
        print("----------------------------")

# --- Fin de la definición de la clase ---

# Creando el objeto impresora1
print("Configuración de Impresora 1")
impresora1 = Impresora("", "", "")
impresora1.leerdatos()
impresora1.mostrardatos()

# Creando el objeto impresora2
print("\nConfiguración de Impresora 2")
impresora2 = Impresora("", "", "")
impresora2.leerdatos()
impresora2.mostrardatos()

print("Software actualizado por Santiago Pacheco")
```
